Linear regression



We'll create a model that predicts crop yields for apples and oranges (target variables) by looking at the average temperature, rainfall, and humidity (input variables or features) in a region. 

Training data:

In a linear regression model, each target variable is estimated to be a weighted sum of the input variables, offset by some constant, known as a bias :

yield_apple  = w11 * temp + w12 * rainfall + w13 * humidity + b1

yield_orange = w21 * temp + w22 * rainfall + w23 * humidity + b2


The learning part of linear regression is to figure out a set of weights w11, w12,... w23, b1 & b2 using the training data, to make accurate predictions for new data. 

The learned weights will be used to predict the yields for apples and oranges in a new region using the average temperature, rainfall, and humidity for that region.

We'll train our model by adjusting the weights slightly many times to make better predictions, using an optimization technique called gradient descent.


```python
import numpy as np
import torch
```

# **Training data**

We can represent the training data using two matrices: inputs and targets, each with one row per observation, and one column per variable.

```python
# Input (temp, rainfall, humidity)
# we can either convert the array to float by explicitly defining dtype ='float32' or just by making one number as float (the entire array will become float) Just put 73.
inputs = np.array([[73, 67, 43], 
                   [91, 88, 64], 
                   [87, 134, 58], 
                   [102, 43, 37], 
                   [69, 96, 70]], dtype='float32')
```

```python
# Targets (apples, oranges)
targets = np.array([[56, 70], 
                    [81, 101], 
                    [119, 133], 
                    [22, 37], 
                    [103, 119]], dtype='float32')
```

Let's convert the arrays to PyTorch tensors.

```python
# Convert inputs and targets to tensors
inputs = torch.from_numpy(inputs)
targets = torch.from_numpy(targets)
print(inputs)
print(targets)
```

# **Linear regression model from scratch**

The weights and biases (w11, w12,... w23, b1 & b2) can also be represented as matrices, initialized as random values. Biases are added as a constant coz if somehow w11 * temp + w12 * rainfall + w13 * humidity all these three are 0. The yield is always a number. There cannot be 0 apple and oranges yield even in extremely worst conditions. It's just to compensate.

The first row of w (i.e. w11, w12, w13) and the first element of b (b1) are used to predict the first target variables.

```python
# Choosing random weights and biases to begin with. randn will choose values that are normally distributed and create a matrix of specified shape. Here w = 2 rows, 3 columns and b = 2  is a vector with 2 values

w = torch.randn(2, 3, requires_grad=True)
b = torch.randn(2, requires_grad=True)
print(w)
print(b)
```

The model is simply a function that performs a matrix multiplication of the inputs and the weights w (transposed always) and adds the bias b (for every observation).

image.png

If you do the matrix multiplications above (5x3) X (3x2) we will get a (5x2) matrix. Then add the (5x2) + (5x2) matrix to get the final (5x2) matrix.

Which should be something like below:

**[ 73w11 + 67w12 + 43w13 + b1 _____________  73w21 + 67w22 + 43w23 + b2 ]**

The first value in the first column is the prediction for Apples, and first value in the second column is the prediction for Oranges, given the input values for first row in the dataset.

```python
# Matrix multiplication can be performed as below: '@' is the symbol in pytorch for matrix multiplication, and the " .t " method returns the transpose of a matrix (tensor).
# Generate predictions
pred = inputs @ w.t() + b
print(pred)
```

```python
# Actual targets
print(targets)
```

There is a big difference between the model's predictions and the actual values. Obviously, with random weights and biases, we can't expect model to just work accurately.

# **Model Evaluation**

With such small data we can visually compare the values and tell that the model is performing poorly. But that is not the right way. We need a mathematical way to evaluate our model. 

We can compare the model's predictions with the actual targets using the **Mean Squared Error (MSE)**

*   Calculate the difference between the two matrices (preds and targets).
*   Square all elements of the difference matrix to remove negative values.
*   Calculate the average of the elements in the resulting matrix.










.

```python
# let's define MSE calculation first. 
# torch.sum returns the sum of all the elements in a tensor. The .numel method of a tensor returns the number of elements in a tensor. Don't just use 5 because we have 10 elements. The no. of elements can change with more data. So we use 'numel()'
def mse(t1, t2):
    diff = t1 - t2
    return torch.sum(diff * diff) / diff.numel()
```

```python
# Compute loss
loss = mse(pred, targets)
print(loss)
```

On average, each element in the prediction differs from the actual target by the square root of the loss. i.e. square root of 53178 ~ 230. And thats pretty bad. (this number will be different every time we run the code)

## **Let's improve this loss and minimize the error**

```python
# to compute loss import .backward(). Without this you will not get the value for print(w.grad) below
loss.backward()
```

```python
print(w)
print(w.grad)
```

# **Adjust weights and biases to reduce the loss**
The loss is a quadratic function of our weights and biases, and our objective is to find the set of weights & biases where the loss is the lowest. 


```python
print(w)
print(w.grad)
```

we have to subtract w - w.grad. If we subtract -1.5026 - (-33003.9219) the values are big. So we multiply the gradients with a very small number (10^-5 in this case) to ensure that we don't modify the weights by a very large amount. We want to take a small step in the downhill direction of the gradient, not a giant leap. This number is called the **learning rate** of the algorithm.

We use **torch.no_grad** to indicate to PyTorch that we shouldn't track, calculate, or modify gradients while updating the weights and biases.



```python
with torch.no_grad():
    w -= w.grad * 1e-5 #this is same as w = w - (w.grad * 1e-5)
    b -= b.grad * 1e-5
```

```python
# check the new values of w and 
print(w)
print(b)
```

```python
# Let's verify that the loss is actually lower
loss = mse(pred, targets)
print(loss)
```

```python
# Before we proceed, we reset the gradients to zero by calling the .zero_() method. We need to do this because PyTorch accumulates gradients. Otherwise, the next time we invoke .backward on the loss, the new gradient values are added to the existing gradients, which may lead to unexpected results.
w.grad.zero_()
b.grad.zero_()
print(w.grad)
print(b.grad)
```

# **Train the model using the above gradient descent theory**
As seen above, we reduce the loss and improve our model using the gradient descent optimization algorithm. Thus, we can train the model using the following steps:


1.   Generate predictions
2.   Calculate the loss
3.   Compute gradients w.r.t the weights and biases
4.   Adjust the weights by subtracting a small quantity proportional to the     
     gradient
5.   Reset the gradients to zero

```python
# Generate predictions
pred = inputs @ w.t() + b
print(pred)
```

```python
# Calculate the loss
loss = mse(pred, targets)
print(loss)
```

```python
# Compute gradients
loss.backward()
print(w.grad)
print(b.grad)
```

```python
# Adjust weights & reset gradients
with torch.no_grad():
    w -= w.grad * 1e-5
    b -= b.grad * 1e-5
    w.grad.zero_()
    b.grad.zero_()
    print(w)
    print(b)
```

```python
# Calculate loss. You will see the loss will reduce.
pred = inputs @ w.t() + b
loss = mse(pred, targets)
print(loss)
```

We have already achieved a significant reduction in the loss just by adjusting the weights and biases slightly using gradient descent.

# **Train for multiple epochs**
To reduce the loss further, we can repeat the process of adjusting the weights and biases using the gradients multiple times. Each iteration is called an epoch. Let's train the model for 100 epochs.

```python
# Train for 100 epochs
for i in range(100):
    pred = inputs @ w.t() + b
    loss = mse(pred, targets)
    loss.backward()
    with torch.no_grad():
        w -= w.grad * 1e-5
        b -= b.grad * 1e-5
        w.grad.zero_()
        b.grad.zero_()
```

```python
# Calculate the new loss
pred = inputs @ w.t() + b
loss = mse(pred, targets)
print(loss)
```

The loss is now much lower than its initial value. Let's look at the model's predictions and compare them with the targets.

```python
print(pred)
print(targets)
```

The predictions are now quite close to the target variables. You can get even better results by training for a few more epochs. Although it is not necessary that more epochs will give you better results. It may also lead to overfitting.
